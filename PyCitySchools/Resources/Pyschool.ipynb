```python
# Dependencies and Setup
import pandas as pd
import os 

# File to Load 
#school_data_to_load = "Users/carlosalvarez/Desktop/Boot/4 pandas/Hmw/schools_complete.csv"
#student_data_to_load = "Users/carlosalvarez/Desktop/Boot/4 pandas/Hmw/students_complete.csv"
csvpathschools = os.path.join('..', 'Hmw', 'schools_complete.csv')
csvpathstudents = os.path.join('..', 'Hmw', 'students_complete.csv')

# Read School and Student Data File and store into Pandas DataFrames
school_data= pd.read_csv(csvpathschools)
student_data = pd.read_csv(csvpathstudents)

# Combine the data into a single dataset.  
df_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
#ADDING THREE COLUMNS: PASS MATH SCORE, PASS READING SCORE, PASS BOTH
#ASSUMPTION: PASS => 70
df_complete["pass math"]= df_complete["math_score"] >=70
df_complete["pass reading"]= df_complete["reading_score"] >=70
df_complete["pass both"]= (df_complete["math_score"] >=70) & (df_complete["reading_score"] >=70)
df_complete.head()
```

```python
# SECTION II DISTRICT SUMMARY
#GET GENERAL INFORMATION ABOUT SCHOOLS
```

```python
#total schools
total_schools= len(df_complete['school_name'].unique())

#total students
total_students= len(df_complete['student_name'])

#total budget
tb= df_complete[["school_name","budget"]].groupby("school_name").mean()
total_budget=tb.sum() 

#average math schore
ave_math_score =df_complete['math_score'].mean()

#average reading schore
ave_reading_score =df_complete['reading_score'].mean()

#   % passing math
#Assumption: not pass is <70
pmath =df_complete[df_complete["pass math"]== True].count()
pass_math_students=pmath.mean()
mathPercentage_pass= pass_math_students/total_students

#   % passing reading
#Assumption: not pass is <70
preading =df_complete[df_complete["pass reading"]== True].count()
pass_reading_students=preading.mean()
readPercentage_pass= pass_reading_students/total_students

# Overall passing
pall =df_complete[df_complete["pass both"]== True].count()
pall_students=pall.mean()
Overall_passing= pall_students/total_students
```

```python
# grouby and create new table for performing data
SN=df_complete.groupby(["school_name"])
SN_new= SN[["math_score","reading_score","budget","size"]].mean() 
SN_type=SN["type"].unique()
new_merge= pd.merge(SN_new,SN_type, how="left", on=["school_name"])
#add column budget per student
new_merge["budget per student"]= new_merge["budget"]/ new_merge["size"]

# Get students by school who pass math, reading, both

#Pass math
Passsm= df_complete.loc[df_complete["pass math"]== True]
Passm_schools=Passsm.groupby(["school_name"])
Pass_math_per_school=Passm_schools["pass math"].count()

#Pass read
Passsr= df_complete.loc[df_complete["pass reading"]== True]
Passr_schools=Passsr.groupby(["school_name"])
Pass_reading_per_school=Passr_schools["pass reading"].count()

#Pass both
Passsboth= df_complete.loc[df_complete["pass both"]== True]
Passboth_schools=Passsboth.groupby(["school_name"])
Pass_both_per_school=Passboth_schools["pass both"].count()
Pass_both_per_school.head()



#MERGE ALL THE DATA IN ONE TABLE
#Merge all tables
new_mergeII= pd.merge(Pass_math_per_school,Pass_reading_per_school, how="left", on=["school_name"])
new_mergeIII= pd.merge(new_mergeII,Pass_both_per_school, how="left", on=["school_name"])

final_merge= pd.merge(new_merge,new_mergeIII, how="left", on=["school_name"])


#add
final_merge["Per_Pass_Math"]= final_merge["pass math"]/final_merge["size"]
final_merge["Per_Pass_Reading"]= final_merge["pass reading"]/final_merge["size"]
final_merge["Per_Pass_Both"]= final_merge["pass both"]/final_merge["size"]

final_merge

#Overall Passing:

#Pass bth
#Passsb= df_complete.loc[df_complete["pass both"]== True]
#Passb_schools=Passsb.groupby(["school_name"])
#Passsb.count()
#Pass_both_per_school=Passr_schools["pass both"].count()

#rename columns
```
